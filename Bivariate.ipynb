```python
import matplotlib.pyplot as plt
plt.scatter(dataset['area'], dataset['price'])
plt.xlabel('Area')
plt.ylabel('Price')
plt.show()
```

```python
import seaborn as sns

plt.figure(figsize=(20, 12))
plt.subplot(2,3,1)
sns.boxplot(x = 'mainroad_yes', y = 'price', data = dataset)
plt.subplot(2,3,2)
sns.boxplot(x = 'guestroom_yes', y = 'price', data = dataset)
plt.subplot(2,3,3)
sns.boxplot(x = 'basement_yes', y = 'price', data = dataset)
plt.subplot(2,3,4)
sns.boxplot(x = 'hotwaterheating_yes', y = 'price', data = dataset)
plt.subplot(2,3,5)
sns.boxplot(x = 'airconditioning_yes', y = 'price', data = dataset)
plt.subplot(2,3,6)
sns.boxplot(x = 'furnishingstatus_unfurnished', y = 'price', data = dataset)
plt.show()
```

```python
import seaborn as sns
correlation_matrix = dataset[["price", "area", "bedrooms", "bathrooms"]].corr()
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm")
plt.show()
```

```python
sns.barplot(x="furnishingstatus_unfurnished", hue="guestroom_yes", data=dataset)
plt.show()
```
